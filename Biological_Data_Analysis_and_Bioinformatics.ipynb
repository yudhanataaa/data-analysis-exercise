## Python Logic

```python
#Contoh Looping dan If/Else : Blood Sample Analyzer
# 1. Siapkan datanya (List)
data_leukosit = [9000, 12000, 8500, 15000, 9800]

# 2. Nyalakan Loop
for jumlah in data_leukosit:

    # 3. Masukkan Logika (If/Else)
    if jumlah > 10000:
        print(f"Sampel {jumlah}: Bahaya")
    else:
        print(f"Sampel {jumlah}: Aman")
```

```python
#Contoh append
# 1. DATA AWAL (Semua sampel masuk)
semua_sampel = [9000, 12000, 8500, 15000, 9800]

# 2. SIAPKAN KERANJANG KOSONG (Penting!)
sampel_bahaya = []

# 3. MULAI LOOPING
for jumlah in semua_sampel:
    if jumlah > 10000:
        sampel_bahaya.append(jumlah)

# 4. LIHAT HASIL AKHIR
print("Daftar Sampel Berbahaya:", sampel_bahaya)
```

```python
#contoh logika and, or, not
suhu_badan = 39
batuk = True

# Memakai 'and' untuk mengecek DUA syarat sekaligus
if suhu_badan > 38 and batuk == True:
    print("Status: Suspect Flu")
else:
    print("Status: Observasi Biasa")
```

```python
#contoh while looping
ph_larutan = 1 #Sangat asam

# Ulangi TERUS selama pH masih di bawah 7
while ph_larutan < 7:
    print(f"pH saat ini {ph_larutan}. Menambahkan basa...")
    ph_larutan = ph_larutan + 1 #Naikkan pH

print("Selesai! Larutan sudah netral.")
```

```python
#Contoh break & continue
sampel_list = ["B", "C", "A", "D", "E"]

for kode in sampel_list:
    if kode == "A":
        print("DITEMUKAN: Sampel A!")
        break #Langsung stop, tidak perlu cek D dan E.

    print(f"Mengecek sampel {kode}...")
```

```python
#Contoh Elif
ph = 7

if ph < 7:
    print("Sifat: Asam")
elif ph == 7:
    print("Sifat: Netral")
else:
    print("Sifat: Basa")
```

```python
#contoh logika bertingkat
rak_sampel = ["Rak A", "Rak B"]
nomor_tabung = [1, 2, 3]

for rak in rak_sampel:           # Loop Luar
    for tabung in nomor_tabung:  # Loop Dalam
        print(f"Mengecek {rak} - Tabung {tabung}")
```

```python
#contoh try & except
data_pasien = [100, "Data Rusak", 50]

for data in data_pasien:
    try:
        # Coba bagi 1000 dengan data
        hasil = 1000 / data
        print(f"Hasil hitungan: {hasil}")
    except:
        # Jika error (karena teks tidak bisa buat pembagian), lari ke sini
        print("Error: Data tidak valid, dilewati.")
```

```python
import pandas as pd
import random

# 1. PERSIAPAN DATA (Generator Otomatis)
# Kita buat list kosong dulu
id_sampel = []
sekuens_dna = []
panjang_dna = []

# Kita lakukan looping 1000 kali untuk bikin data palsu
for i in range(1000):
    # Bikin ID (misal: S-0, S-1, dst)
    id_sampel.append(f"S-{i}")

    # Bikin DNA acak (Pilih A/T/C/G secara random sebanyak 10 huruf)
    # (Ini trik python lanjutan, anggap saja mesin pembuat DNA)
    dna_acak = "".join(random.choices(["A", "T", "C", "G"], k=10))
    sekuens_dna.append(dna_acak)

    # Hitung panjangnya
    panjang_dna.append(len(dna_acak))

# 2. BUNGKUS KE DATAFRAME
data_genom = {
    "ID_Sampel": id_sampel,
    "Sekuens": sekuens_dna,
    "Panjang_Bp": panjang_dna
}

df_genom = pd.DataFrame(data_genom)

# 3. INSPEKSI DATA (Cek Kualitas)
print("=== 5 Data Teratas (Head) ===")
display(df_genom.head())

print("\n=== Ringkasan Data (Info) ===")
# .info() sangat penting untuk melihat tipe data dan memori
df_genom.info()
```

## Missing Data Imputation in Public Health Datasets

```python
import pandas as pd

# 1. Baca data
df = pd.read_csv('life_data.csv')

# 2. Intip isinya (Perhatikan kolom GDP dan Life expectancy)
print("--- Sampel Data WHO ---")
df.head()

# 3. DIAGNOSA: Hitung berapa data yang hilang di setiap kolom
print("\n--- Laporan Data Hilang (Missing Values) ---")
# sort_values supaya kita bisa lihat yang paling banyak hilangnya di paling atas
df.isnull().sum().sort_values(ascending=False)
```

```python
df.head()
```

```python
#MENGISI TABEL KOSONG DENGAN FILLNA
# 1. Menghitung Rata-rata GDP dari data yang ADA
rata_rata_gdp = df['GDP'].mean()
print("Rata-rata GDP Dunia (Dollar):", rata_rata_gdp)

# 2. EKSEKUSI: Mengisi (Fill) data kosong di kolom GDP dengan nilai rata-rata tadi
# inplace=False (default) artinya kita harus simpan balik ke variabel df['GDP']
df['GDP'] = df['GDP'].fillna(rata_rata_gdp)

# 3. VERIFIKASI: Mari kita cek apakah 'lubang' di GDP sudah tertutup?
print("\n--- Cek Ulang Data Hilang di GDP ---")
print("Jumlah missing value di GDP sekarang:", df['GDP'].isnull().sum())
```

```python
# 1. Pilih hanya kolom yang isinya angka (Float/Integer) karena kita mau isi pakai Rata-rata
kolom_angka = df.select_dtypes(include=['float64', 'int64']).columns

# 2. Loop (Perulangan) otomatis: Isi setiap kolom angka dengan rata-ratanya masing-masing
# Kode ini membaca: "Untuk setiap kolom di dalam daftar kolom_angka, isi yang kosong dengan rata-rata kolom itu"
for kolom in kolom_angka:
    df[kolom] = df[kolom].fillna(df[kolom].mean())

# 3. FINAL CHECK: Apakah dataset sudah 100% bersih?
print("--- Sisa Data Hilang (Total) ---")
print(df.isnull().sum().sum())
# Harusnya hasilnya 0 (Kecuali kolom teks seperti 'Status' negara)
```

```python
#CONTOH HISTOGRAM & KDE (Univariate Analysis)
plt.figure(figsize=(10, 6))

# sns.histplot: Membuat histogram
# kde=True: Menambahkan garis kurva mulus (Kernel Density Estimate)
# bins=30: Membagi data menjadi 30 kelompok balok
sns.histplot(data=df, x='charges', kde=True, color='teal', bins=30)

plt.title('Distribusi Tagihan Medis (Charges)', fontsize=16)
plt.xlabel('Besar Tagihan ($)', fontsize=12)
plt.ylabel('Frekuensi (Jumlah Orang)', fontsize=12)

plt.show()
```

```python
#CONTOH SCATTER PLOT (Multivariate Analysis)
plt.figure(figsize=(12, 7))

# SCATTER PLOT AJAIB
# x: Kita taruh BMI di sumbu mendatar
# y: Kita taruh Tagihan di sumbu tegak
# hue: Kita minta Seaborn bedakan warna berdasarkan kolom 'smoker'
# palette: Pilihan palet warna (misal: 'magma', 'coolwarm', 'viridis')

sns.scatterplot(data=df, x='bmi', y='charges', hue='smoker',
                palette='coolwarm', s=60, alpha=0.8)

plt.title('Hubungan BMI vs Tagihan Medis (Dibedakan Status Perokok)', fontsize=16)
plt.xlabel('Body Mass Index (BMI)', fontsize=12)
plt.ylabel('Tagihan ($)', fontsize=12)

plt.show()
```

```python
#CONTOH BOXPLOT (membandingkan kategori)
plt.figure(figsize=(12, 6))

# BOX PLOT ESTETIK
# x: Wilayah (Kategori)
# y: Tagihan (Angka)
# hue: Kita pisahkan lagi berdasarkan Perokok/Bukan
sns.boxplot(data=df, x='region', y='charges', hue='smoker',
            palette='Set2') # Palette 'Set2' warnanya pastel lembut

plt.title('Perbandingan Biaya Medis per Wilayah (Smoker vs Non-Smoker)', fontsize=16)
plt.ylabel('Tagihan ($)', fontsize=12)
plt.xlabel('Wilayah', fontsize=12)

# Memindahkan legenda supaya tidak menutupi gambar
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import io
from google.colab import files

# --- 1. UPLOAD FILE (Pilih 'bisnis.csv') ---
print("Silakan upload file 'bisnis.csv' yang sudah kamu buat...")
uploaded = files.upload()

# Baca file
nama_file = next(iter(uploaded))
df = pd.read_csv(io.BytesIO(uploaded[nama_file]))

print("\n✅ Data Berhasil Masuk!")
display(df.head())

# --- 2. EDA: HEATMAP KORELASI ---
# Kita ingin menjawab: FAKTOR APA YANG BIKIN PROFIT NAIK?
plt.figure(figsize=(8, 6))

# Hitung korelasi hanya untuk kolom angka
korelasi = df.select_dtypes(include=['number']).corr()

# Gambar Heatmap
sns.heatmap(korelasi, annot=True, cmap='RdBu', center=0, vmin=-1, vmax=1)
plt.title('Apa yang Mempengaruhi Profit?', fontsize=14)
plt.show()

# --- 3. SCATTER PLOT PEMBUKTIAN ---
plt.figure(figsize=(12, 5))

# Marketing vs Profit
plt.subplot(1, 2, 1)
sns.regplot(data=df, x='Marketing Spend', y='Profit', color='orange')
plt.title('Marketing vs Profit')

# R&D vs Profit
plt.subplot(1, 2, 2)
sns.regplot(data=df, x='R&D Spend', y='Profit', color='blue')
plt.title('R&D (Inovasi) vs Profit')

plt.tight_layout()
plt.show()
```

```python
#GENERATE DATA
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. MEMBUAT DATASET SINTETIS (1500 Baris) ---
np.random.seed(42)
n_per_species = 500 # 500 baris per spesies

# A. Bakteri Acidophilus (Suka Asam, Air Keruh)
acid = pd.DataFrame({
    'pH': np.random.normal(loc=4.5, scale=0.5, size=n_per_species), # pH Rendah
    'Temperature': np.random.normal(loc=25, scale=3, size=n_per_species),
    'Turbidity': np.random.normal(loc=80, scale=10, size=n_per_species), # Keruh
    'Species': 'Acidophilus'
})

# B. Bakteri Thermophilus (Suka Panas, pH Netral)
thermo = pd.DataFrame({
    'pH': np.random.normal(loc=7.0, scale=0.5, size=n_per_species),
    'Temperature': np.random.normal(loc=55, scale=5, size=n_per_species), # Panas!
    'Turbidity': np.random.normal(loc=40, scale=10, size=n_per_species),
    'Species': 'Thermophilus'
})

# C. Bakteri Neutralis (Suka Air Bersih & Sejuk)
neutral = pd.DataFrame({
    'pH': np.random.normal(loc=7.2, scale=0.4, size=n_per_species),
    'Temperature': np.random.normal(loc=20, scale=2, size=n_per_species), # Sejuk
    'Turbidity': np.random.normal(loc=10, scale=5, size=n_per_species), # Jernih
    'Species': 'Neutralis'
})

# Gabungkan semuanya jadi satu dataset raksasa
df = pd.concat([acid, thermo, neutral])

# Acak urutannya biar seperti data asli
df = df.sample(frac=1).reset_index(drop=True)

# Simpan jadi CSV (Pura-pura ini file hasil sensor)
df.to_csv('water_quality.csv', index=False)

print(f"✅ File 'water_quality.csv' berhasil dibuat dengan {df.shape[0]} baris data!")
display(df.head())
```

```python
# --- 1. CEK KESEIMBANGAN DATA ---
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Species', palette='Set2')
plt.title('Jumlah Sampel per Spesies Bakteri', fontsize=14)
plt.show()

# --- 2. ANALISIS PERBEDAAN (BOX PLOT) ---
# Kita ingin melihat karakteristik pH dan Suhu tiap bakteri
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot Kiri: Perbandingan pH
sns.boxplot(data=df, x='Species', y='pH', palette='viridis', ax=axes[0])
axes[0].set_title('Karakteristik Keasaman (pH) per Spesies')
axes[0].grid(True, linestyle='--', alpha=0.3)

# Plot Kanan: Perbandingan Suhu
sns.boxplot(data=df, x='Species', y='Temperature', palette='magma', ax=axes[1])
axes[1].set_title('Karakteristik Suhu per Spesies')
axes[1].grid(True, linestyle='--', alpha=0.3)

plt.show()

# --- 3. PETA TOTAL (PAIR PLOT) ---
# Ini visualisasi paling keren untuk klasifikasi!
# Kita bisa melihat pemisahan spesies berdasarkan semua kombinasi sensor
sns.pairplot(df, hue='Species', palette='bright')
plt.show()
```

```python
#ANSWER CHALLENGE 1
def hitung_diskon(jumlah_barang):
    harga_per_barang = 10000
    total_awal = jumlah_barang * harga_per_barang
    diskon = 0

    if jumlah_barang > 10:
        diskon = 0.20
        print("Dapat Diskon 10%")
    elif jumlah_barang > 5:
        diskon = 0.10
        print("Dapat Diskon 20%")
    else:
        diskon = 0
        print("Tidak ada diskon")

    total_akhir = total_awal - (total_awal * diskon)
    return total_akhir

tagihan_budi = hitung_diskon(12)
print(f"Total yang harus dibayar Budi: Rp {tagihan_budi:,.0f}")
```

```python
#ANSWER CHALLENGE 9
dna_asli = "ATGC"

def cari_pasangan_dna(urutan):
    # 1. Buat Kamus Pasangan (Peta)
    kamus_pasangan = {
        'A': 'T',
        'T': 'A',
        'G': 'C',
        'C': 'G'
    }

    # 2. Siapkan Wadah Kosong (String kosong)
    dna_baru = ""

    # 3. Loop satu per satu hurufnya
    for basa in urutan:
        # Cari pasangannya di kamus, lalu masukkan ke wadah
        pasangan = kamus_pasangan[basa]
        dna_baru = dna_baru + pasangan

    return dna_baru

# --- TES HASIL ---
print(f"DNA Asli : {dna_asli}")
print(f"Pasangan : {cari_pasangan_dna(dna_asli)}")
```

```python
#REVIEW BASIC BIOINFORMATICS
# --- 1. DATA INPUT ---
dna_kotor = "atgGccTtcGGaTaaGCc"
print(f"1. DNA Mentah  : {dna_kotor}")

# --- 2. PRE-PROCESSING (Cleaning & Transkripsi) ---
# Langkah A: Bersihkan (Uppercase)
dna_bersih = dna_kotor.upper()

# Langkah B: Ubah ke RNA (T jadi U)
rna = dna_bersih.replace('T', 'U')
print(f"2. RNA Sequence: {rna}")

# --- 3. TRANSLASI (Logic Core) ---
kamus_protein = {
    'AUG': 'Methionine',
    'GCC': 'Alanine',
    'UUC': 'Phenylalanine',
    'GGA': 'Glycine',
    'UAA': 'STOP'
}

protein_list = [] # Wadah untuk menyimpan hasil terjemahan

# LOOPING 3 HURUF (STEP 3)
# Kita melompat setiap 3 karakter: 0, 3, 6, 9...
for i in range(0, len(rna), 3):
    # Slicing: Ambil 3 huruf (disebut Kodon)
    kodon = rna[i : i+3]

    # Cek apakah potongan itu benar 3 huruf? (Jaga-jaga kalau sisa di akhir)
    if len(kodon) < 3:
        break

    # Terjemahkan pakai Kamus
    if kodon in kamus_protein:
        nama_protein = kamus_protein[kodon]

        # Cek kondisi STOP
        if nama_protein == 'STOP':
            print(f"   >> Menemukan Kodon STOP ({kodon}). Berhenti.")
            break # Keluar dari loop paksa

        # Masukkan ke wadah
        protein_list.append(nama_protein)
    else:
        print(f"   >> Peringatan: Kodon {kodon} tidak ada di kamus!")

# --- 4. HASIL AKHIR ---
# Gabungkan list jadi string cantik dengan tanda sambung '-'
hasil_akhir = "-".join(protein_list)
print(f"3. Hasil Protein: {hasil_akhir}")
```

```python
#REVIEW BASIC BIOINFORMATICS
#MINI DNA TRANSLATOR

import time

# --- 1. DATABASE (KAMUS GENETIK) ---
# Ini adalah tabel kodon standar.
# Kiri = Kodon RNA, Kanan = Simbol Asam Amino (1 Huruf)
# M = Start (Methionine), _ = STOP
GENETIC_CODE = {
    'AUG': 'M', 'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L',
    'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S', 'UAU': 'Y',
    'UAC': 'Y', 'UAA': '_', 'UAG': '_', 'UGU': 'C', 'UGC': 'C',
    'UGA': '_', 'UGG': 'W', 'CUU': 'L', 'CUC': 'L', 'CUA': 'L',
    'CUG': 'L', 'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'CAU': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q', 'CGU': 'R',
    'CGC': 'R', 'CGA': 'R', 'CGG': 'R', 'AUU': 'I', 'AUC': 'I',
    'AUA': 'I', 'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
    'AAU': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K', 'AGU': 'S',
    'AGC': 'S', 'AGA': 'R', 'AGG': 'R', 'GUU': 'V', 'GUC': 'V',
    'GUA': 'V', 'GUG': 'V', 'GCU': 'A', 'GCC': 'A', 'GCA': 'A',
    'GCG': 'A', 'GAU': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
    'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'
}

# --- 2. FUNGSI LOGIKA UTAMA ---

def bersihkan_dna(dna_mentah):
    """Membersihkan spasi dan mengubah ke huruf besar"""
    return dna_mentah.upper().replace(" ", "").replace("\n", "")

def transkripsi_dna_ke_rna(dna):
    """Mengubah T menjadi U"""
    return dna.replace('T', 'U')

def translasi_rna_ke_protein(rna):
    """Menerjemahkan RNA 3 huruf sekaligus menjadi Protein"""
    protein_chain = []

    # Loop dengan step 3 (Membaca Kodon)
    for i in range(0, len(rna), 3):
        kodon = rna[i : i+3]

        # Jika sisa huruf kurang dari 3, abaikan
        if len(kodon) < 3:
            break

        # Cari di kamus. Jika tidak ada, kasih tanda '?'
        asam_amino = GENETIC_CODE.get(kodon, '?')

        if asam_amino == '_': # Tanda STOP
            print(f"   [INFO] Menemukan Kodon STOP ({kodon}).")
            break

        protein_chain.append(asam_amino)

    return "-".join(protein_chain)

# --- 3. USER INTERFACE (MAIN PROGRAM) ---
def jalankan_software():
    print("="*40)
    print("  BIOINFORMATICS TOOL: DNA TRANSLATOR  ")
    print("="*40)
    print("Petunjuk: Masukkan urutan DNA mentah.")
    print("Contoh: atg gcc ttt gaa taa")
    print("-" * 40)

    # A. INPUT USER
    input_user = input(">> Masukkan DNA: ")

    print("\n[Sedang Memproses...]")
    time.sleep(1) # Simulasi loading biar keren

    # B. PROSES
    dna_bersih = bersihkan_dna(input_user)
    rna = transkripsi_dna_ke_rna(dna_bersih)
    protein = translasi_rna_ke_protein(rna)

    # C. OUTPUT LAPORAN
    print("\n" + "="*40)
    print("LAPORAN ANALISIS")
    print("="*40)
    print(f"1. Panjang DNA : {len(dna_bersih)} basa")
    print(f"2. Struktur RNA: {rna}")
    print("-" * 40)
    print(f"3. RANTAI PROTEIN:\n     {protein}")
    print("="*40)

# Jalankan Programnya!
jalankan_software()
```
